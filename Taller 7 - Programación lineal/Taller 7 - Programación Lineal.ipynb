<h2>Punto 1</h2>

<i>[Taha, 2007] Ozark Farms use at least 800 lb of special feed daily. The special feed is a mixure of corn and soybean meal with the following composition:

|Feed stuff   | Protein| Fiber| Cost ($/lb)|
|-------------|------------|------------|------------|
|Corn   |.09| .02| .30|
|Soybean meal |.60|.06| .90|

The dietary requeriments of special feed are at least 30% protein and at most 5% fiber. 
Determine the daily minimum-cost feed mix.</i>

Variables:
$$x_1=Corn$$
$$x_2=Soybean meal$$

Función:
$$Z(min)=0.3x_1+0.9x_2$$

Restricciones:
\begin{align}
x_1+x_2 &\geq 800\\
0.09x_1+0.60x_2 &\geq 0.3(x_1+x_2)\\
0.02x_1+0.06x_2 &\leq 0.05(x_1+x_2)\\
x_1,x_2 &\geq 0\\
\end{align}

    Por lo tanto
\begin{align}
x_1+x_2 &\geq 800\\
0.21x_1-0.30x_2 &\leq 0\\
0.03x_1-0.01x_2 &\geq 0\\
x_1,x_2 &\geq 0\\
\end{align}

Solución:

```python
fig, ax = plt.subplots(figsize=(8, 8))
x1 = np.linspace(0, 1700)
y1 = 800 - x1
y2 = 0.7*x1
y3 = 3*x1

# Primera restricción
plt.plot(x1, y1, lw=3, label=r'$x_1+x_2 \geq 800$')

# Segunda  restricción
plt.plot(x1, y2 , lw=3, label=r'$0.21x_1-0.30x_2 \leq 0$')

# Tercera  restricción
plt.plot(x1, y3, lw=3, label=r'$0.03x_1-0.01x_2 \geq 0$')

# Constantes positivos
plt.plot(np.zeros_like(x1), x1, lw=3, label=r'$x_2\geq 0$')
plt.plot(x1, np.zeros_like(x1), lw=3, label=r'$x_1\geq 0$')

# Función para el valor minimo
plt.plot(x1, 486.2713-0.33*x1, '.', lw=3, label=r'$Z(min)=0.3x_1+0.9x_2$')

#Región
plt.fill_between(x1, np.maximum(y1,y2), np.maximum(y1,y3), alpha=0.1)

#Costo mínimo
plt.plot([470.59], [329.41], 'o', color='red')
plt.text(471, 330, u'Costo mínimo', color='red')

plt.xlabel(r'$x_1$', fontsize=16)
plt.ylabel(r'$x_2$', fontsize=16)
plt.xlim(-0.1, 1600)
plt.ylim(-0.1, 1600)
plt.legend(fontsize=14)
plt.show()
```

<h3>Punto 2</h3>

<i>Bridgeway Company manufactures a printer and keyboard. The contribution margins of the printer and keyboard are 30 dollars and 20 dollars, respectively. Two types of skilled labor are required to manufacture these products: soldering and assembling. A printer requires 2 hours of soldering and I hour of assembling. A keyboard requires 1 hour of soldering and 1 hour of assembling. Bridgeway has 1,000 soldering hours and 800 assembling hours available per week. There are no constraints on the supply of raw materials. Demand for keyboards is unlimited, but at most 350 printers are sold each week. Bridgeway wishes to maximize its weekly total contribution margin.</i>
<br><i>Draw the feasible region and compute the objective function on the intersections.</i></br>

Tabla:

|Manufacture   | Printer | Keyboard | Cost ($/unit) | Hours available (h/week) |
|-------------|------------|------------|------------|------------|
|Soldering   |2 h|1 h|30|1000|
|Assembling |1 h|1 h|20|800|
|<i>Sold max</i>|350

Variables:
$$x_1=Soldering$$
$$x_2=Assembling$$

Función:

$$Z(max)=30x_1+20x_2$$

Restricciones:
\begin{align}
x_1+x_2 &\leq 800\\
2x_1+x_2 &\leq 1000\\
x_1&\leq 350\\
x_1,x_2 &\geq 0\\
\end{align}

Solución:

```python
fig, ax = plt.subplots(figsize=(8, 8))
x1 = np.linspace(0, 1200)
y1 = 800 - x1
y2 = 1000 - (2*x1)
y3 = np.ones_like(x1)*350

# Primera restricción
plt.plot(x1, y1, lw=3, label=r'$x_1+x_2 \leq 800$')

# Segunda  restricción
plt.plot(x1, y2 , lw=3, label=r'$2x_1+x_2 \leq 1000$')

# Tercera  restricción
plt.plot(y3, x1, lw=3, label=r'$x_1 \leq 350$')

# Constantes positivos
plt.plot(np.zeros_like(x1), x1, lw=3, label=r'$x_2\geq 0$')
plt.plot(x1, np.zeros_like(x1), lw=3, label=r'$x_1\geq 0$')

# Función para el valor minimo
plt.plot(x1, 900-1.5*x1, '.', lw=3, label=r'$Z(max)=30x_1+20x_2$')

#Región
plt.fill_between(x1, 0, np.minimum(y1,y2), x1<=y3, alpha=0.1)

#Costo mínimo
plt.plot([200], [600], 'o', color='red')
plt.text(200, 600, u'Máximo producción', color='red')

plt.xlabel(r'$x_1$', fontsize=16)
plt.ylabel(r'$x_2$', fontsize=16)
plt.xlim(-0.1, 1100)
plt.ylim(-0.1, 1100)
plt.legend(fontsize=14)
plt.show()
```

<i>1 .  Describe the dual version of the problem.
\begin{align}
\max &  \; x_1+x_2\\
\text{s.t. } & x_1+2x_2 &\leq 4\\
 & 4x_1+2x_2 &\leq 12\\
 & -x_1+x_2 &\leq 1\\
 & x_1,x_2 \geq 0\\
\end{align}</i>

1. Para hallar la ecuación Z(min), los coeficientes de esto serían los limites que tiene las restricciónes del problema. Despues de esto se agregan las variables respectivas ($y_1, y_2, y_3...$).
    <br><b>Restricciones iniciales:</b></br>
    <br>$x_1+2x_2 \leq$ <font color="red">4</font></br>
    <br>$4x_1+2x_2 \leq$ <font color="blue">12</font></br>
    <br>$-x_1+x_2 \leq$ <font color="green">1</font></br>
    <br><b>Por lo tanto:</b></br>
    <br>$Z(min)=$<font color="red"> 4</font>$y_1+$<font color="blue"> 12</font>$y_2+$<font color="green">1</font>$y_3$</br>
2. Para obtener las funciones de las restricciones de esto, se debe transponer los coeficientes de las restricciones de z(max) en forma vertical, y despues agregar las variables respectivas. Por lo tanto si se tenia restricción de 3x2 se convierte a ser restricción de 2x3.
    <br><b>Restricciones iniciales:</b></br>
    <br><font color="red">1</font>$x_1+$<font color="blue">2</font>$x_2 \leq$ 4</br>
    <br><font color="red">4</font>$x_1+$<font color="blue">2</font>$x_2 \leq$ 12</br>
    <br><font color="red">-1</font>$x_1+$<font color="blue">1</font>$x_2 \leq$ 1</br>
    <br><b>Por lo tanto:</b></br>
    <br>(1) <font color="red">1</font>$y_1+$<font color="red">4</font>$y_2+($<font color="red">-1</font>$)y_3$</br>
    <br>(2) <font color="blue">2</font>$y_1+$<font color="blue">2</font>$y_2+$<font color="blue">1</font>$y_3$</br>
3. Para reconocer los límites de cada restricciones, primero se mira los coeficientes de la ecuacion Z(max) cual estos serán los valores de límites. Despues se mira si los límites de las valiables originale $(x_1, x_2)$ si es "maximo" o "minimo" ya que estos límites seran para estas restricciones. En el caso de que no tienen límites, se pone el "=".
    <br><b>Ecuacion inicial:</b></br>
    <br>$Z(max)=$<font color="red"> 1</font>$x_1+$<font color="blue"> 1</font>$x_2$</br>
    <br>$x_1$<font color="green">$\geq$</font>$0$</br>
    <br>$x_2$<font color="pink">$\geq$</font>$0$</br>
    <br><b>Por lo tanto:</b></br>
    <br>(1) $1y_1+4y_2-y_3$<font color="green">$\geq$</font><font color="red"> 1</font></br>
    <br>(2) $2y_1+2y_2+y_3$<font color="pink">$\geq$</font><font color="blue"> 1</font></br>

    Construcción del problema dual:

\begin{align}
\min &  \; 4y_1+12y_2+y_3\\
\text{s.t. }
 & y_1+4y_2 - y_3 &\geq 1\\
 & 2y_1+2y_2+y_3 &\geq 1\\

\end{align}</i>

<i>2 . Solve the dual problem using the graphical method or simplex method.</i>

```python
x = np.outer(np.linspace(-1, 2, 50), np.ones(50))
y = x.copy().T
z = x.copy().T
z1 = x+4*y
z2 = -2*x-2*y
z3 = 0

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_zlabel('$x_3$')

ax.plot_surface(x, y, z1, color='red', alpha=0.7)
ax.plot_surface(x, y, z2, color='yellow', alpha=0.7)
```
